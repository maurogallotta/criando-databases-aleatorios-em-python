## Primeira Forma de Gerar Dados Aleatorios

```python
import csv
import random
from datetime import datetime, timedelta
import pandas as pd
```

```python
# Marca(Mostra) as versões dos pacotes carregados no jupyter notebook
%reload_ext watermark
%watermark -a "Mauro Gallotta" --iversions
```

```python
# Gerar dados de exemplo
def generate_data(num_rows):
    data = []
    for i in range(num_rows):
        # Gerar ID de transação único
        id_transacao = i + 1

        # Gerar horário de pedido aleatório nos últimos 365 dias
        horario_pedido = datetime.now() - timedelta(days=random.randint(0, 365))

        # Gerar localidade em números aleatórios (supondo números de 1 a 100)
        localidade_em_numeros = random.randint(1, 100)

        # Gerar nome(número) do item aleatório
        nome_item = f"Item_{random.randint(1, 10)}"

        # Gerar quantidade do item (supondo entre 1 e 10)
        quantidade_item = random.randint(1, 10)

        # Gerar latitude e longitude aleatórias (supondo valores em uma faixa razoável)
        latitude = round(random.uniform(-90, 90), 6)
        longitude = round(random.uniform(-180, 180), 6)

        data.append([id_transacao, horario_pedido, localidade_em_numeros, nome_item, quantidade_item, latitude, longitude])

    return data

# Escrever dados em um arquivo CSV
def write_to_csv(data, filename):
    with open(filename, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['id_transacao', 'horario_pedido', 'localidade_em_numeros', 'nome_item', 'quantidade_item', 'latitude', 'longitude'])
        writer.writerows(data)

# Número de linhas desejadas no dataset
num_linhas = int(input('Digite quantidade de linhas do seu Database? '))
num_rows = num_linhas

# Gerar dados das linhas
data = generate_data(num_rows)

# Escrever(grava) dados em um arquivo CSV
write_to_csv(data, 'dataset.csv')
print("Seu Database esta pronto!")
```

## Segunda Forma de Gerar Dados Aleatorios com a biblioteca "Faker"

### Listagem de termos quem pode ajudar a geração de algumas informações

faker: Assim que o script for instalado em seu ambiente, em desenvolvimento você poderia usar python -m faker instead

-h, --help: Mostra as mensagem de ajuda

--version: Mostra o número da versão instalada

-o FILENAME: Redireciona a saída para o nome de arquivo especificado

-l {bg_BG,cs_CZ,...,zh_CN,zh_TW}: Permite o uso de um provedor localizado

-r REPEAT: Irá gerar um número especificado de saídas

-s SEP: Irá gerar o separador especificado após cada saída gerada

-i {my.custom_provider other.custom_provider} Lista de provedores personalizados adicionais a serem usados. Observe que esse é o caminho de importação do pacote que contém sua classe Provider, e não a própria classe Provider personalizada..

fake: É o nome do falso para o qual gerar uma saída, como nome, endereço ou texto

[fake argument ...] argumentos opcionais para passar para o falso (por exemplo, o perfil falso recebe uma lista opcional de nomes de campos separados por vírgula como primeiro argumento)

```python
import csv
from faker import Faker

# Crie uma instância do Faker com localização em pt_BR
fake = Faker('pt_BR')
```

```python
# Marca(Mostra) as versões dos pacotes carregados no jupyter notebook
%reload_ext watermark
%watermark -a "Mauro Gallotta" --iversions
```

```python
# Semente de instancia, para o exemplo ficar igual em qualquer maquina(Otimo para teste)
fake.seed_instance(54)
# 'Leonardo Jesus'

# Exemplo de uso de alguns métodos para gerar dados fictícios
nome_completo = fake.name()
primeiro_nome = fake.first_name()
ultimo_nome = fake.last_name()
endereco = fake.address()
bairro = fake.neighborhood()
cep = fake.postcode()
cidade = fake.city()
nome_estado = fake.estado_nome()
sigla_estado = fake.estado_sigla()
telefone = fake.phone_number()
email = fake.email()
data_de_nascimento = fake.date_of_birth()
rg = fake.rg()
cpf = fake.ssn()
trabalho = fake.job()
empresa = fake.company()
cnpj = fake.cnpj()
placa_do_carro = fake.license_plate()
alfanumerico = fake.vin()
texto = fake.text()

# Exibindo os dados gerados
print("Nome Completo:", nome_completo)
print(" ")
print("Primeiro Nome:", primeiro_nome)
print(" ")
print("Último Nome:", ultimo_nome)
print(" ")
print("Endereço:", endereco)
print(" ")
print("Bairro" , bairro)
print(" ")
print("Cep:", cep)
print(" ")
print("Cidade:", cidade)
print(" ")
print("Estado Nome:", nome_estado)
print(" ")
print("Estado Sigla:", sigla_estado)
print(" ")
print("Telefone:", telefone)
print(" ")
print("E-mail:", email)
print(" ")
print("Data de Nascimento:", data_de_nascimento)
print(" ")
print("RG:", rg)
print(" ")
print("CPF:", cpf)
print(" ")
print("Trabalho:", trabalho)
print(" ")
print("Empresa:", empresa)
print(" ")
print("Cnpj:", cnpj)
print(" ")
print("Placa do Carro:", placa_do_carro)
print(" ")
print("Código Alfanumérico:", alfanumerico)
print(" ")
print("Texto Aleatório:", texto)
print(" ")
```

```python
registros_a_gerar = int(input('Quantos registros quer gerar? '))
registros_gerados = 0

# Gerar registros fictícios
registros = []
for i in range(registros_a_gerar):
    registro = {
        'ID': i + 1,
        'Nome': fake.name(),
        'Endereco': fake.address(),
        'Bairro': fake.neighborhood(),
        'Cep': fake.postcode(),
        'Cidade': fake.city(),
        'Estado Nome': fake.estado_nome(),
        'Estado Sigla': fake.estado_sigla(),
        'Telefone': fake.phone_number(),
        'Email': fake.unique.email(),
        'Data de Nascimento': fake.date_of_birth(),
        'RG': fake.unique.rg(),
        'CPF': fake.ssn(),
        'Trabalho': fake.job(),
        'Empresa': fake.company(),
        'Cnpj': fake.cnpj(),
        'Placa de Carro': fake.license_plate(),
        'Codigo Alfanumerico': fake.vin(),
        'Comentario Aleatorio': fake.text(),
    }
    registros.append(registro)
    registros_gerados +=1

# Nome do arquivo CSV de saída
nome_arquivo_csv = 'registros_aleatorios.csv'

# Escreva os registros no arquivo CSV
with open(nome_arquivo_csv, mode='w', newline='') as arquivo_csv:
    campos = ['ID', 'Nome', 'Endereco', 'Bairro', 'Cep', 'Cidade', 'Estado Nome', 'Estado Sigla',  'Telefone', 'Email', 'Data de Nascimento', 'RG', 'CPF', 'Trabalho', 'Empresa', 'Cnpj', 'Placa de Carro', 'Codigo Alfanumerico', 'Comentario Aleatorio']
    escritor_csv = csv.DictWriter(arquivo_csv, fieldnames=campos)

    escritor_csv.writeheader()
    for registro in registros:
        escritor_csv.writerow(registro)

print(f'Foram gerados {registros_gerados} / {registros_a_gerar} registros, e exportados para o arquivo "{nome_arquivo_csv}".')
```
